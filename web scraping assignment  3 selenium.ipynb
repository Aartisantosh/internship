```python
for i in page_url:
    driver.get(i)
    time.sleep(5)
    
    try:
        name = driver.find_elements(By.XPATH, '//div[@class="property-name"]')  # scraping hostel name
        # Assuming you want to append the text of the first element found
        hostel_name.append(name[0].text if name else '-')
    except NoSuchElementException:
        hostel_name.append('-')
    

    try:
        name = driver.find_elements(By.XPATH,'//div[@class="property-name"]') # scraping  hostel name
        hostel_name.append(i.text)
    except NoSuchElementException:
        hostel_name.append('-')
    

 #scraping privates dorm 4 bed from price
    try:
        pvt_price = driver.find_element(By.XPATH,"//a[@class='prices']//div[1]//div")
        pvt_prices.append(pvt_price.text)
    except NoSuchElementException:
        pvt_prices.append('-')
   

            
    # scraping dorms from price
    try:
        dorms = driver.find_element(By.XPATH,"//a[@class='prices']//div[2]/div")
        dorms_price.append(dorms.text)
    except NoSuchElementException:
        dorms_price.append('-')
            
    

    # scraping ratings
    try:
        rat = driver.find_element(By.XPATH,'//div[@class="flex-20"]/div[1]/div[1]')
        rating.append(rat.text)
    except NoSuchElementException:
        rating.append('-')
        
        
    # scraping total review
    try:
        rws = driver.find_element(By.XPATH,'//div[@class="flex-20"]/div[1]/div[2]/div[2]')
        reviews.append(rws.text.replace('Total Reviews',''))
    except NoSuchElementException:
        reviews.append('-')
        
        
    # fetching over all review
    try:
        overall = driver.find_element(By.XPATH,'//div[@class="flex-20"]/div[1]')
        over_all.append(overall.text)
    except NoSuchElementException:
        over_all.append('-')
        
        
    # fetching property description
    try:
        disc = driver.find_element(By.XPATH,'//div[@class="flex-80"]/div[4]/div/div[2]')
        description.append(disc.text)
    except NoSuchElementException:
        over_all.append('-')
    
    # scraping facilities
    try:
        wifi = driver.find_elements(By.XPATH,'//div[@id="facilities-section"]/div/ul/li[1]/ul/li[3]')
        facilities.append(wifi.text)
    except NoSuchElementException:
        facilities.append('-')
        
        
#     # do click on show more button for description
#     try:
#         driver.find_element(By.XPATH,"//a[@class='toggle-content']").click()
#         time.sleep(4)
#     except NoSuchElementException:
#         pass
    
    
        
```
